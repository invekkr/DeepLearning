````python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten
from keras import Sequential
from keras.datasets import mnist
````

````python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
````

````python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))


model.summary()
````


# ğŸ“Œ CNN Model Explanation (Layer by Layer)


## ğŸ§  1ï¸âƒ£ First Convolution Layer

* **32 filters** â†’ output will have 32 feature maps
* **Kernel size = (3,3)**
* **Padding = 'valid'** â†’ no padding (size decreases)
* **Input shape = (28,28,1)** â†’ grayscale image

### Output Size Calculation

Formula (stride = 1, no padding):

[
Output = N - F + 1
]

[
28 - 3 + 1 = 26
]

So output becomes:

```
(26, 26, 32)
```

---

## ğŸ§  2ï¸âƒ£ Second Convolution Layer

Input: (26, 26, 32)

[

]

Output:

```

```

---

## ğŸ§  3ï¸âƒ£ Third Convolution Layer

Input: (24, 24, 32)

[
24 - 3 + 1 = 22
]

Output:

```
(22, 22, 32)
```

---

## ğŸ”¥ Size Shrinking

```
28 â†’ 26 â†’ 24 â†’ 22
```

Because:

* Padding = 'valid'
* No zero padding added
* Spatial dimensions reduce after each convolution

---

## ğŸ§± 4ï¸âƒ£ Flatten Layer

Converts:

```

```

[

Output shape:

```

```

---

## ğŸ§  5ï¸âƒ£ Dense Layer

## ğŸ¯ 6ï¸âƒ£ Output Layer

# ğŸ“Œ Overall Flow

1. Input image (28Ã—28Ã—1)
2. Three convolution layers extract features
3. Spatial size shrinks
4. Flatten converts feature maps to vector
5. Dense layers perform classification




````python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))

model.summary()
````


# ğŸ§  Key Differences from Previous Model

* `padding='same'` â†’ Output size is controlled (does not shrink due to filter size).
* `strides=(2,2)` â†’ Filter moves 2 steps at a time.
* Stride reduces spatial size.
* Here, shrinking happens because of **stride**, not filter size.

---

# ğŸ“ Output Size Calculation (Very Important)

For SAME padding:

[
\text{Output size} = \left\lceil \frac{N}{S} \right\rceil
]

Where:

* N = input size
* S = stride

---

## 1ï¸âƒ£ First Convolution Layer

Input:

```
(28, 28, 1)
```

Stride = 2

[
28 / 2 = 14
]

Output:

```
(14, 14, 32)
```

---

## 2ï¸âƒ£ Second Convolution Layer

Input:

```
(14, 14, 32)
```

[

```

```

---

## 3ï¸âƒ£ Third Convolution Layer

Input:

```
(7, 7, 32)
```

[
7 / 2 = 3.5 \Rightarrow 4 \text{ (ceiling)}
]

```
(4, 4, 32)
```

---

# ğŸ”¥ Observe the Size Change

```
28 â†’ 14 â†’ 7 â†’ 4
```

Here:

* Shrinking happens because of **stride = 2**
* Padding keeps edges protected
* No extra shrink from kernel size
