````python
def step(z):
    return 1 if z>0 else 0
````


### What this function does

```python
def step(z):
    return 1 if z > 0 else 0
```

This function is called a **step activation function** (also known as a **threshold function**).

It converts a **numerical input** into a **binary class output**.

---

### Why this function is needed

In the perceptron, we compute:

```text
z = w₀ + w₁x₁ + w₂x₂
```

* `z` can be **any real number** (positive or negative)
* But for **binary classification**, we need a **class label**

  * `0` or `1`

The step function performs this conversion.

---

### How it works (decision rule)

* If `z > 0` → return `1`
* If `z ≤ 0` → return `0`

This creates a **hard decision boundary**.

In words:

> “If the weighted sum is positive, predict class 1, otherwise class 0.”

---

### Geometric intuition (very important)

* The equation

  ```text
  w₀ + w₁x₁ + w₂x₂ = 0
  ```

  represents a **straight line**

* This line divides the space into two regions:

  * One side → `z > 0` → class 1
  * Other side → `z ≤ 0` → class 0

The step function decides **which side of the line** a point lies on.

---

### Why it’s called an activation function

* It decides whether the neuron **“fires” (1)** or **“does not fire” (0)**
* Mimics early ideas of biological neurons

---

### Major limitation (very important for Deep Learning)

The step function is **not differentiable** at `z = 0`.

Because of this:

* We **cannot compute gradients**
* **Backpropagation does not work**
* It cannot be used in deep neural networks

That’s why modern DL uses:

* Sigmoid
* ReLU
* Tanh

---

### Why it is still used here

* Perceptron learning **does not use gradients**
* It relies on **mistake-driven updates**
* So the step function works perfectly

---

### One-line summary

> The step function turns a weighted sum into a hard class decision by checking which side of the decision boundary the input lies on.




````python
def perceptron(x,y):
    x = np.insert(x,0,1,axis=1)
    weights = np.ones(x.shape[1])
    lr = 0.1
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(x[j],weights))
        weights = weights + lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
````

Perceptron Function

This function implements a **Perceptron**, which is the earliest and simplest form of a neural network. Understanding this code deeply is extremely important because **every deep learning model is an extension of this idea**.

I’ll explain **what**, **why**, and **how** for every single part.

---

1. Overall goal of this function

The goal of this function is to:

* Learn how to separate two classes (0 and 1)
* Using a **straight line** (linear decision boundary)
* By adjusting weights based on mistakes

In simple words:

> The perceptron tries to find a line that correctly divides the data into two groups.

---

2. Input to the function

def perceptron(x, y):

* `x` → feature matrix
  Shape before bias: (100, 2)
  Each row = one data point
  Each column = one feature

* `y` → labels
  Shape: (100,)
  Values are either 0 or 1

So each data point looks like:
(x₁, x₂) → y

---

3. Adding the bias term

x = np.insert(x, 0, 1, axis=1)

WHY is this needed?

A linear model is:
w₀ + w₁x₁ + w₂x₂ = 0

* w₀ is the **bias**
* Without bias, the line is forced to pass through the origin
* Bias gives the model **flexibility to shift the line**

WHAT this line does:

* Inserts a column of 1s at the beginning of `x`

Before:
[x₁, x₂]

After:
[1, x₁, x₂]

Now:

* `1` multiplies with bias weight w₀
* x₁ multiplies with w₁
* x₂ multiplies with w₂

New shape of `x`:
(100, 3)

---

4. Weight initialization

weights = np.ones(x.shape[1])

WHAT this means:

* One weight per input feature
* Since x has 3 columns → weights has 3 values

weights = [w₀, w₁, w₂]

WHY initialize with ones?

* Any small constant works
* Learning happens through updates, not initial values
* Avoids zero-gradient situations

---

5. Learning rate

lr = 0.1

WHAT learning rate does:

* Controls how big each weight update is

If lr is:

* Too small → learning is very slow
* Too large → learning becomes unstable

0.1 is a reasonable middle value for simple datasets.

---

6. Training loop

for i in range(1000):

WHAT this means:

* The model updates weights 1000 times
* Each iteration tries to correct mistakes

Important:

* Perceptron does NOT converge in one pass
* It needs repeated exposure to data

---

7. Random data point selection (Stochastic Learning)

j = np.random.randint(0, 100)

WHY random?

* Picks one random data point at a time
* This is **Stochastic Gradient Descent (SGD)**

WHY SGD instead of full dataset?

* Faster updates
* Better generalization
* Mimics real deep learning training

At this point:

* We are training on ONE sample, not all 100

---

8. Computing the prediction

y_hat = step(np.dot(x[j], weights))

Let’s break this completely.

a) Dot product

np.dot(x[j], weights)

This computes:
w₀·1 + w₁·x₁ + w₂·x₂

This value is called **z** (net input).

b) Step activation function

step(z):

* If z ≥ 0 → output 1
* If z < 0 → output 0

WHY step function?

* This is a **binary classifier**
* Output is a hard decision, not probability

IMPORTANT:

* Modern DL replaces this with sigmoid or ReLU
* Step function is not differentiable

---

9. Understanding prediction result

y_hat is the model’s guess:

* 0 or 1

Compare it with true label y[j]:

* If equal → prediction is correct
* If different → model made a mistake

---

10. Weight update rule (MOST IMPORTANT PART)

weights = weights + lr * (y[j] - y_hat) * x[j]

This single line is the **heart of learning**.

Let’s break it fully.

a) Error term

(y[j] - y_hat)

Possible values:

* 0 → correct prediction
* +1 → predicted 0 but true label is 1
* -1 → predicted 1 but true label is 0

b) Why multiply by x[j]?

* Updates weights proportional to input values
* Features that contributed more get larger correction

c) Why multiply by learning rate?

* Controls step size
* Prevents overshooting

---

11. What happens in different cases?

Case 1: Correct prediction
y = y_hat
→ Error = 0
→ No weight update

Case 2: False negative (missed positive)
y = 1, y_hat = 0
→ Error = +1
→ Weights increase → pushes boundary toward positive class

Case 3: False positive
y = 0, y_hat = 1
→ Error = -1
→ Weights decrease → pushes boundary away

This is how the perceptron **learns from mistakes**.

---

12. Return statement

return weights[0], weights[1:]

Meaning:

* weights[0] → bias
* weights[1:] → slope coefficients

---

13. Final learned model

After training, the perceptron has learned:

w₀ + w₁x₁ + w₂x₂ = 0

This equation represents:

* A straight line in 2D
* Decision boundary separating two classes

---



````python
intercept_, coef_ = perceptron(x,y)
print(intercept_, coef_)
````

### What this code does

```python
intercept_, coef_ = perceptron(x, y)
print(intercept_, coef_)
```

* Calls the `perceptron` function to **train the model**
* Extracts the learned **bias (intercept)** and **feature weights (coefficients)**
* Prints them to see the final learned parameters

---

### What `perceptron(x, y)` returns

Inside your perceptron function you return:

```python
return weights[0], weights[1:]
```

So:

* `weights[0]` → **intercept (bias)**
* `weights[1:]` → **coefficients for input features**

That’s why unpacking works:

```python
intercept_, coef_ = perceptron(x, y)
```

---

### Meaning of `intercept_`

* This is the **bias term** ( w_0 )
* It shifts the decision boundary up/down (or left/right)
* Without it, the line must pass through the origin

Mathematically:

```text
z = intercept_ + coef_[0]·x₁ + coef_[1]·x₂
```

---

### Meaning of `coef_`

* `coef_` is an array of feature weights
* For 2 features:

```text
coef_ = [w₁, w₂]
```

Each weight controls:

* Direction of the decision boundary
* Importance of that feature

---

### What `print(intercept_, coef_)` shows you

* The **final learned parameters** after training
* These values define the separating line

Example output (varies every run):

```text
-0.8 [2.3 0.1]
```

---

### How this relates to the decision boundary

The perceptron learns the line:

```text
w₀ + w₁x₁ + w₂x₂ = 0
```

Using returned values:

```text
intercept_ + coef_[0]·x₁ + coef_[1]·x₂ = 0
```

This equation is what separates the two classes.

---

### Important note

Because:

* You use random sampling (`np.random.randint`)
* Training is stochastic

👉 The printed values **will change each time you run the code**

This is expected and normal.

---

### One-line intuition

> This code prints the exact numbers that define the line learned by the perceptron.




````python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
````



### What this code is doing

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

This code converts the **perceptron’s learned weights** into the **slope–intercept form of a straight line**:

```text
y = m·x + b
```

---

### Where this comes from (most important part)

Your perceptron learned the decision boundary equation:

```text
intercept_ + coef_[0]·x₁ + coef_[1]·x₂ = 0
```

Let:

* x₁ → x (horizontal axis)
* x₂ → y (vertical axis)

So the equation becomes:

```text
intercept_ + coef_[0]·x + coef_[1]·y = 0
```

---

### Rearranging to y = mx + b

Move terms:

```text
coef_[1]·y = -coef_[0]·x - intercept_
```

Now divide everything by `coef_[1]`:

```text
y = -(coef_[0] / coef_[1])·x - (intercept_ / coef_[1])
```

---

### Matching with your code

So:

```python
m = -(coef_[0] / coef_[1])   # slope
b = -(intercept_ / coef_[1]) # y-intercept
```

✔ Exactly matches the math above.

---

### Meaning of `m` (slope)

* Controls the **tilt of the decision boundary**
* Positive `m` → upward sloping line
* Negative `m` → downward sloping line
* Larger magnitude → steeper line

---

### Meaning of `b` (intercept)

* Where the line crosses the **y-axis**
* Shifts the boundary up or down
* Comes from the bias term learned by the perceptron

---

### Why this is useful

Once you have `m` and `b`, you can:

* Plot the decision boundary
* Visually verify if the perceptron learned correctly
* Understand how the model separates classes

Example plotting logic:

```text
y = m·x + b
```

---

### One-line intuition

> You are converting the perceptron’s internal weights into a human-readable straight line.




````python
x_inp = np.linspace(-3,3,100)
y_inp = m*x_inp+b
````

This step is about **creating points to draw the decision boundary line**.

### What this code does

```python
x_inp = np.linspace(-3, 3, 100)
y_inp = m * x_inp + b
```

It generates **100 points on the decision boundary line** so that the line can be plotted smoothly.

---

### Line 1: Creating x-values

```python
x_inp = np.linspace(-3, 3, 100)
```

* `np.linspace(start, end, num)`
* Generates `num` evenly spaced values between `start` and `end`

So this creates:

* 100 x-values
* Starting from `-3`
* Ending at `+3`

Why `-3` to `3`?

* Your data roughly lies in this range
* The line should cover the full plot area

Think of this as:

> “Pick many x-values so we can draw a smooth line.”

---

### Line 2: Computing corresponding y-values

```python
y_inp = m * x_inp + b
```

This uses the **line equation**:

```text
y = m·x + b
```

* `m` → slope (learned from perceptron)
* `b` → intercept (learned from perceptron)
* `x_inp` → array of x-values

NumPy automatically applies this formula to **all 100 x-values at once** (vectorization).

So:

* For every `x` in `x_inp`
* Compute the corresponding `y`
* Store it in `y_inp`

---

### What you have now

* `x_inp` → x-coordinates of the decision boundary
* `y_inp` → y-coordinates of the decision boundary

Together:

```text
(x_inp[i], y_inp[i]) lies on the decision boundary
```

---

### Why this is necessary

Mathematically, a line is infinite.
But to **draw** it:

* We need discrete points
* Many points → smooth-looking line

This is exactly what this code does.

---

### Deep Learning intuition

* The perceptron learned **numbers** (`m` and `b`)
* This step turns those numbers into **geometry**
* You are literally visualizing what the model learned

---

### One-line intuition

> You are generating points so the perceptron’s learned decision boundary can be plotted as a straight line.



````python
plt.plot(x_inp,y_inp, color='red', linewidth=2)
plt.scatter(x[:,0],x[:,1],c=y,cmap = "plasma",s=70)
plt.ylim(-3,2)
````
